# RT extraction walkthrough

Five steps:
1. parse a Presentation `.log` file
2. pair each probe Picture with the next valid button press → per-trial RTs
3. summarize per subject
4. plot the distribution + per-trial scatter
5. handle the BIDS `events.tsv` variant

```python
from rt_demo.parsers.presentation import parse_log
from rt_demo.extract import extract_rts
from rt_demo.summary import summarize
from rt_demo.plots import plot_rt_distribution, plot_rt_over_trials

log = parse_log('synthetic_data/sub-01_task-demo_run-1.log')
log.head()
```

```python
rts = extract_rts(log)
rts.insert(0, 'subject', 'sub-01')
rts.head(10)
```

```python
summarize(rts, group_cols=['subject'])
```

```python
plot_rt_distribution(rts, title='sub-01 RT distribution');
```

```python
plot_rt_over_trials(rts, title='sub-01 RT across trials');
```
